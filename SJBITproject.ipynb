```python
controlnet = ControlNetModel.from_pretrained(
    "lllyasviel/control_v11p_sd15_canny",
    torch_dtype=torch.float16
)

pipe = StableDiffusionControlNetImg2ImgPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    controlnet=controlnet,
    torch_dtype=torch.float16
)

pipe = pipe.to("cuda")
```

```python
import torch
import cv2
import numpy as np
from PIL import Image

print("✅ Imports ready")
```

```python
image_path = list(uploaded.keys())[0]

image = cv2.imread(image_path)

# Safety check (avoid crash)
if image is None:
    raise ValueError("Image not loaded properly")

image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Resize (VERY IMPORTANT for SD)
image = cv2.resize(image, (512, 512))

input_image = Image.fromarray(image)

input_image
```

```python
edges = cv2.Canny(image, 100, 200)

# Strengthen edges (better reconstruction)
edges = cv2.dilate(edges, np.ones((3,3), np.uint8), iterations=1)

edges = np.stack([edges]*3, axis=-1)

edges_pil = Image.fromarray(edges)

edges_pil
```

```python
prompt = "Imagine how it would have been in past as indian monuments and regenerate the missing parts with clean structure,architectural design, highly detailed, photorealistic, 4k, Ignore the background and do not generate any nature"

negative_prompt = "blurry, distorted, broken, unrealistic, messy, cartoon"
output = pipe(
    prompt=prompt,
    negative_prompt=negative_prompt,
    image=input_image,
    control_image=edges_pil,
    strength=0.6,
    num_inference_steps=35,
    guidance_scale=7.5,
    controlnet_conditioning_scale=0.9
).images[0]

output
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))

# Original Image
plt.subplot(1,2,1)
plt.imshow(input_image)
plt.title("Original (Ruins)")
plt.axis("off")

# Reconstructed Image
plt.subplot(1,2,2)
plt.imshow(output)
plt.title("Reconstructed Output")
plt.axis("off")

plt.show()
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.imshow(input_image)
plt.title("Input Image")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(edges_pil)
plt.title("Edge Map (Canny)")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(output)
plt.title("Final Reconstruction")
plt.axis("off")

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def sharpness(img):
    img = np.array(img)
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    return cv2.Laplacian(gray, cv2.CV_64F).var()

original_score = sharpness(input_image)
output_score = sharpness(output)

plt.bar(["Original", "Reconstructed"], [original_score, output_score])

plt.title("Sharpness Improvement")
plt.ylabel("Score")

plt.show()
```

```python
from skimage.metrics import structural_similarity as ssim

orig = cv2.cvtColor(np.array(input_image), cv2.COLOR_RGB2GRAY)
gen = cv2.cvtColor(np.array(output), cv2.COLOR_RGB2GRAY)

score, _ = ssim(orig, gen, full=True)

print("SSIM Score:", score)
```

```python
def psnr(img1, img2):
    mse = np.mean((img1 - img2) ** 2)
    if mse == 0:
        return 100
    return 20 * np.log10(255.0 / np.sqrt(mse))

psnr_value = psnr(orig, gen)

print("PSNR:", psnr_value)
```

```python
import time

start = time.time()

_ = pipe(
    prompt=prompt,
    image=input_image,
    control_image=edges_pil,
    strength=0.6
)

end = time.time()

print("Inference Time:", end - start, "seconds")
```
